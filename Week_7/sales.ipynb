```python
import pandas as pd
```

```python
df = pd.read_csv("raw_sales.csv")
df.reset_index(inplace=True)
df
```

```python
df.columns
```

```python
df.rename(columns={'index':"prod_id",'  ("prod_id"':"qty",' "qty"':"price",' "price"':"discount",' "discount"':"member", ' "member"':"txn_id", ' "txn_id"':"start_txn_time",' "start_txn_time")':'NaN'},inplace=True)
df
```

```python
df.drop(['NaN'], axis=1,inplace=True)
```

```python
df["prod_id"] = df["prod_id"].apply(lambda x: x[4:-1])
df["qty"] = df["qty"].apply(lambda x: x [2:-1])
df["price"] = df["price"].apply(lambda x: x[2:-1]) 
df["discount"] = df["discount"].apply(lambda x: x[2:-1]) 
df["member"] = df["member"].apply(lambda x: x[2:-1])
df["txn_id"] = df["txn_id"].apply(lambda x: x[2:-1])
df["start_txn_time"] = df["start_txn_time"].apply(lambda x: x[2:-2])
```

```python
df
```

Using regular expressions to find (')

```python
import re
for i in range(len(df["start_txn_time"])):
    if matches := re.search(r"^(.+)'$",df["start_txn_time"][i]):
        print(i, matches.groups())
        df["start_txn_time"][i] = matches.group(1)
```
